```python
import numpy as np
import matplotlib.pyplot as plt
from iminuit import Minuit
from iminuit.cost import ExtendedBinnedNLL
from scipy.stats import norm

Vam = [10, 14, 19, 23, 28, 33, 38, 43, 47, 52, 56, 61]
Vpulse = [15, 17, 20, 23, 26, 30, 34, 38, 41, 44, 48, 52, 57, 62, 66, 71]

for l in range(len(Vam)):
    raw_data = np.loadtxt(r"C:\Users\lenovo\Desktop\AlfaRuth24-25_S1\calibrazione_Vbias\900mm2_241Am\Vbias_900mm2_241Am_06-11-24_V" + str(Vam[l])+".dat")
    
    data1 = []
    
    for i in range(len(raw_data)):
        for j in range(int(raw_data[i])):
            data1.append(i)
    
    data = np.array(data1)
    
    signal = []
    
   #per 50 Am : 5220/5400
    #per 50pulse: 4819/4860
    for i in range(len(data)):
        if data[i] > 5220 and data[i] < 5400: signal.append(data[i])

    
    #plt.hist(signal, bins=40)
    #plt.show()

    N=sum(raw_data)
    def model(x, N1, N2, mu1, sigma1, mu2, sigma2):
    	return N1*norm.cdf(x, loc = mu1, scale = sigma1) + N2*norm.cdf(x, loc = mu2, scale = sigma2)
    
    
    N1_ext = 0
    N2_ext = 0
    
    for i in range(5275, 5350): #picco1 (più a dx)
        N1_ext = N1_ext + raw_data[i]
    
    for i in range(5225, 5275): #picco2
        N2_ext = N2_ext + raw_data[i]
    
    numbins=[]
    chi2=[]

    for nbins in range(90, 250):
       
        bin_content, bin_edges = np.histogram(signal, bins = nbins)
    
        costf = ExtendedBinnedNLL(bin_content, bin_edges, model)
    
        my_minuit = Minuit(costf, N1 = N1_ext, N2 = N2_ext,  mu1 = 5300, sigma1 = 10, mu2= 5250, sigma2= 10)
        my_minuit.migrad()
        plt.hist(signal)
        # print(my_minuit.valid)
        if ((my_minuit.valid == True) and (my_minuit.fval/my_minuit.ndof > 0.1 )):
            chi2.append(my_minuit.fval/my_minuit.ndof)
            numbins.append(nbins)
#print(numbins[np.array(chi2).argmin()])

    index = np.array(chi2).argmin() #dove c'è il fit migliore

    nbins = numbins[index]
    print(nbins)
    bin_content, bin_edges = np.histogram(signal, bins = nbins)
    
    N_events = sum(bin_content)
    
    costf = ExtendedBinnedNLL(bin_content, bin_edges, model)
    
    my_minuit = Minuit(costf, N1 = N1_ext, N2 = N2_ext,  mu1 = 5300, sigma1 = 10, mu2= 5250, sigma2= 10)
    
    my_minuit.migrad()
    display(my_minuit)
    ch1 = my_minuit.values['mu1']
    err_ch1 = max(1, my_minuit.errors['mu1'])
    fwhm1 = 2.355 * my_minuit.values['sigma1']
    err_fwhm1 = 2.355 * my_minuit.errors['sigma1']
    frac1 = fwhm1 / ch1
    err_frac1 = frac1 * np.sqrt((err_ch1/ch1)**2 + (err_fwhm1/fwhm1)**2)
    
    #salva i dati che ci interessano in myfile.txt
    
    with open("Am241_def_900mm2.txt", "a") as f:
        f.write(f" {Vam[l]} {ch1} {err_ch1} {fwhm1} {err_fwhm1} {frac1} {err_frac1} {my_minuit.fmin.reduced_chi2:.1f}")
        f.write("\n")
```
